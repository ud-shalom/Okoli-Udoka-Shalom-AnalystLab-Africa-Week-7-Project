**EXTRACT (COLLECTING THE DATA)**

**Importing Libraries**

```python
# Cell 1: Import the tools (libraries) we need
import requests      # This tool allows Python to 'talk' to the API website
import pandas as pd  # This tool allows us to organize data into tables
from datetime import datetime # This tool helps us get the current date and time

print("Tools imported successfully. Ready for Step 2!")
```

**Setting the API Key**

```python
# To set API key here
API_KEY = "YOUR_API_KEY_HERE"

print("API Key saved!")
```

**Defining Cities and Base URL**

```python
# Set the cities to track
# We have included Port Harcourt, Accra, and Cape Town
cities = ["Port Harcourt", "Accra", "Cape Town"]

# This is the web address to get the data
base_url = "http://api.openweathermap.org/data/2.5/weather"

# This is an empty list to store the data later
weather_data = []

print("Cities list (Port Harcourt, Accra, Cape Town) is ready!")
```

**The Collection Loop**

```python
# Create the collection loop
for city in cities:
    params = {"q": city, "appid": API_KEY, "units": "metric"}
    response = requests.get(base_url, params=params)

    if response.status_code == 200:
        data = response.json()
        weather_data.append({
            "City": data["name"],
            "Temperature": data["main"]["temp"],
            "Humidity": data["main"]["humidity"],
            "Condition": data["weather"][0]["description"],
            "Wind_Speed": data["wind"]["speed"],
            "Date_Time": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        })
        print(f"Collected: {city}")
```

**TRANSFORM DATA (PROCESSING THE DATA)**

**Creating the Data Table**

```python
# Show the results
print(weather_data)
```

**Displaying the Data for Comparison**

```python
# Create the table
df = pd.DataFrame(weather_data)

# Show the clean table
print(df)
```

```python
# Displaying the data for comparison
# Looking at relevant fields to compare the three cities
print(df[['City', 'Temperature', 'Humidity', 'Condition']])
```

**Identifying Insights**

```python
# Identifying the city with the highest humidity
# Comparing humidity values to determine which city is the most humid
max_humidity_row = df.loc[df['Humidity'].idxmax()]
print(f"City with highest humidity: {max_humidity_row['City']} at {max_humidity_row['Humidity']}%")
```

**LOAD (STORING THE DATA)**

```python
# Task 3: Load Data
# Save the table as a CSV file
df.to_csv("weather_data.csv", index=False)

print("Data saved successfully as 'weather_data.csv'!")
```
